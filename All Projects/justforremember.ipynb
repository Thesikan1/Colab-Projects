```python
import numpy as np
```

```python
a=np.array([1,2,3])
b=np.array([2,4,6])
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

```python
a=np.array([2,3])
b=np.array([4,4])
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

```python
random_matrix=np.random.rand(3,3)
print(random_matrix)
```

```python
random_int=np.random.randint(100,size=(3,3))
print(random_int)
```

```python
random_int1=np.random.randint(200,size=(4,4))
print(random_int1)
```

```python
#matrix with all the values 1
matrix_3=np.ones((2,3),dtype=int)
print(matrix_3)
```

```python
#null matrix or zero matrix
null_matrix=np.zeros((3,3),dtype=int)
print(null_matrix)
```

```python
#identity matrix
identity_matrix=np.identity(3,dtype=int)
print(identity_matrix)
```

```python
#transpose of the matrix
a=np.random.randint(100,size=(3,3))
print(a)
```

```python
transpose_of_a=np.transpose(a)
print(transpose_of_a)
```

```python
b=np.random.randint(200,size=(3,3))
transpose_of_b=np.transpose(b)
print(transpose_of_b)
```

```python
#matrix Addition'
a=np.array([[2,3],[4,6]])
b=np.array([[2,5],[8,9]])
print(a)
print(b)
```

```python
sum=a+b
print(sum)
```

```python
#multiply a matrix by a  scaler
a=5
b=np.random.randint(100,size=(3,3))
print(b)
product=np.multiply(a,b)
print(product)
```
