```python
#Image parameters
img_size=224
batch_size=32
```

Train Test Split

```python
#Image Data Generator
data_gen=ImageDataGenerator(
    rescale=1/255,
    validation_split=0.2 #Use 20% of data for validation
)
```

```python
#Train Generators
train_generator=data_gen.flow_from_directory(
    base_dir,
    target_size=(img_size,img_size),
    batch_size=batch_size,
    subset='training'
)
```
